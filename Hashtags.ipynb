```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./final_df.csv',index_col=0)
```

```python
df.hashtags = df.hashtags.fillna('')
df.hashtags = [x.split(' ') for x in df.hashtags]
```

```python
#중복값 체크
for idx,i in enumerate(df.hashtags):
    if len(set(i)) != len(i):
        print(idx)
```

```python
token_ls = []
for i in df.hashtags:
    for j in i:
        token_ls.append(j)
token_ls.sort()

token_set_ls = []
for i in df.hashtags:
    for j in i:
        if j not in token_set_ls:
            token_set_ls.append(j)
token_set_ls.sort()
```

```python
print('전체 hashtags의 token의 개수:', len(token_ls))
print('hashtags의 unique token의 개수:', len(token_set_ls))
```

## token의 빈도

```python
tmp = np.zeros(len((token_set_ls)))

for idx,i in enumerate(token_set_ls):
    tmp[idx] = token_ls.count(i)  

val_ls = [int(x) for x in tmp]

token_df = pd.DataFrame(val_ls, index = token_set_ls, columns = {'count'})

hashtag_token_df = token_df.sort_values(by = 'count', ascending = False)
```
